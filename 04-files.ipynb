# Working with files

We use one of the standard libraries, glob. It looks into the hard drive and reads files as a list.

```python
import glob

glob.glob('data/inflammation-*.csv')
```

The output is a list, and not in order! The order depends on the OS you're using.
History: in October more than 100 published studies revealed to be wrong bc of the order of the files, for the exact same glitch. For this reason, we NEED a sorter!

```python
import numpy as np
import matplotlib.pyplot as plt

filenames = sorted(glob.glob('data/inflammation-*.csv'))

for file_i in filenames:

    arthritis = np.loadtxt(file_i, delimiter=",")

    fig = plt.figure(figsize=(10, 3))
    
    axes1 = fig.add_subplot(1, 3, 1)  #1 row, 3 columns, number 1
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.set_ylabel('average')
    axes1.plot(arthritis.mean(axis=0))
    
    axes2.set_ylabel("max")
    axes2.plot(arthritis.max(axis=0))

    axes3.set_ylabel("min")
    axes3.plot(arthritis.min(axis=0))

    # one last line to make the plot look better
    fig.tight_layout
    plt.show()
```
